**Q1) Python Program to Check Palindrome Number**

```python
n=int(input("Enter a number:"))
temp=n
rev=0
while(n>0):
    dig=n%10
    rev=rev*10+dig
    n=n//10
if(temp==rev):
    print(f"{temp} is a palindrome!")
else:
    print(f"{temp} is not a palindrome!")
```

**Q3) Python Program to Find the Factorial of a Number**

```python
n = int(input("Enter a number: "))
num = n
if n < 0:
    print(f"Factorial of {num} is 0")
elif n == 0 or n == 1:
    print(f"Factorial of {num} is 1")
else:
    fact = 1
    while(n > 1):
        fact *= n
        n -= 1
    print(f"Factorial of {num} is {fact}")
```

**Q4) Python Program to Print the Fibonacci sequence**

```python
n = int(input("Enter a number: ")) 
n1 = n
l = len(str(n))
Sum = 0
while n != 0:
    r = n % 10
    Sum = Sum+(r**l)
    n = n//10
if n1 == Sum:
    print(f"{n1} is an armstrong number")
else:
    print(f"{n1} is not an armstrong number")
```

**Q1) Write a code to implement Guess Game Guess Game** 
**constraints / Rules** 
**you have 5 chance to guess correct number which is choosen by computer / friend** 
**Range which is allowed to select a number 1-50** 
**After each guess you should give some tips / hints to user to improve their chance of winning** 
**for example if user guess a number which is higher than actual number you should say `hint: think a low number!** 
**if user guess a number which is lower than actual number you should say `hint: think a big number!`** 
**if user wins we will show a congrats messege and stop our game** 
**if any time user guess is out of limit than print a warning message saying Think in Limits 1-50 only!** 
**but we do not count this as a chance**

```python
import random
import math
x = random.randint(1, 50)
print("You have 5 chances to guess the number")
count = 0
print("Guess the number between 1 and 50:")
while count < 5:
    count += 1
    guess = int(input("Guess a number: "))
    if guess < 1 or guess > 50:
        print("Think in limits 1-50 only")
        count -= 1
    if x == guess:
        print(f"Congratulations you did it in {count} tries")
        break
    elif x > guess:
        print("Think of a higher number")
    elif x < guess:
        print("Think of a lower number")
    if count >= 5:
        print(f"\nThe number is {x}")
        print("\tBetter Luck Next time!")
```

**Q5) Write a program to sort a list without using inbuilt functions (list.sort not allowed)** 
    
    a) Using Bubble Sort Algorithm 
    b) Using Insertion Sort Algorithm
    c) Using Selection Sort Algorithm

```python
# Bubble Sort
list1 = []
n = int(input("Enter number of values: "))
for i in range(n):
    value = int(input("Enter a value: "))
    list1.append(value)
for i in range(0,n-1):  
    for j in range(n-i-1):  
        if(list1[j]>list1[j+1]):   
            list1[j],list1[j+1] = list1[j+1], list1[j]  
print(f"The sorted list is: {list1}") 
```

```python
# Insertion Sort
list1 = []
n = int(input("Enter number of values: "))
for i in range(n):
    value = int(input("Enter a value: "))
    list1.append(value)
for i in range(1, n):  
    value = list1[i]  
    j = i - 1  
    while j >= 0 and value < list1[j]:  
        list1[j + 1] = list1[j]  
        j -= 1  
    list1[j + 1] = value  
print(f"The sorted list is: {list1}") 
```

```python
# Selection Sort
list1 = []
n = int(input("Enter number of values: "))
for i in range(n):
    value = int(input("Enter a value: "))
    list1.append(value)
for i in range(n):  
    minIndex = i  
    for j in range(i+1, n):  
        if list1[j] < list1[minIndex]:  
            minIndex = j  
    list1[i], list1[minIndex] = list1[minIndex], list1[i]    
print(f"The sorted list is: {list1}")
```

**Q4) Write a program to read roll number and marks of 10 students in 3 subjects. Calculate total marks of only those students who get more than or equal to 40 marks in each subject. Count the number of students whose total is more than 200. Print the roll number of the students who gets the highest total.**

```python
def get_roll():
    while True:
        roll = int(input("Enter roll number (1000-9999): "))
        if 1000 <= roll <= 9999:
            return roll
        else:
            print("Invalid roll number. Please try again.")

def get_marks():
    while True:
        marks = int(input("Enter marks (0-100): "))
        if 0 <= marks <= 100:
            return marks
        else:
            print("Invalid marks. Please try again.")

total_marks = 0
count_total_more_than_200 = 0
highest_total = 0
highest_roll_number = 0

for _ in range(10):
    print("Enter details for student", _+1)
    roll = get_roll()

    total_subjects = 0
    for _ in range(3):
        marks = get_marks()
        if marks >= 40:
            total_subjects += marks

    if total_subjects >= 120:
        total_marks += total_subjects
        if total_subjects > 200:
            count_total_more_than_200 += 1
        if total_subjects > highest_total:
            highest_total = total_subjects
            highest_roll_number = roll

print("Total marks of students with >= 40 marks in each subject:", total_marks)
print("Number of students with total marks > 200:", count_total_more_than_200)
print("Roll number of student with highest total marks:", highest_roll_number)
```
